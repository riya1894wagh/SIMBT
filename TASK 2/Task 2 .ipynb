1. **Image loading:**

        Load the input image into the program.
    
    
2. **Grayscale conversion:**

        Convert the input image into grayscale format to remove any color information.
    
    
3. **Invert colors:** 

        Invert the grayscale image to obtain the negative image.
    
    
4. **Gaussian blur:** 

        Apply a Gaussian blur to smooth the image and remove any noise.
    
    
5. **Edge detection:** 

        Use an edge detection algorithm (such as the Canny edge detector) to detect the edges in the image.
    
    
6. **Thresholding:** 

        Apply a thresholding operation to the edge-detected image to remove any weak edges and obtain a binary image.
    
    
7. **Dilation:** 

        Dilate the binary image to thicken the edges.
    
    
8. **Final sketch:** 

        Combine the dilated binary image with the negative grayscale image to obtain the final pencil sketch.

## 1. Import   Library

```python

import cv2  
import numpy as np
from skimage import color
from skimage import io
from PIL import Image, ImageOps
from PIL import Image, ImageFilter
import matplotlib.pyplot as plt
```

## 2. Image Loading

```python
 
image = io.imread("shinchan.jpg")
```

```python
plt.imshow(image)
```

## 3. **Grayscale conversion:**

        Convert the input image into grayscale format to remove any color information.
    

```python
imgGray = color.rgb2gray(image)

io.imshow(imgGray)

 
```

## 4. Invert colors:

```python
img_invert = 255 - imgGray
io.imshow(img_invert)
```

## 5. Gaussian blur

```python
Gaussian_img = cv2.GaussianBlur(img_invert,(5,5),cv2.BORDER_DEFAULT)
io.imshow(Gaussian_img)
```

## 6 Invert Blur image

```python
img_invert = 255 - imgGray
io.imshow(Gaussian_img)
```

## 7. Thresholding:

```python
retval, threshold = cv2.threshold(Gaussian_img, 12, 255, cv2.THRESH_BINARY)
io.imshow(Gaussian_img)
```

## 8 Sketch: 

```python
sketch_img=cv2.divide(imgGray,Gaussian_img, scale=256.0)
plt.imshow(sketch_img)
 
```

## 9. Saving image

```python
cv2.imwrite("skech.png", sketch_img)
```
